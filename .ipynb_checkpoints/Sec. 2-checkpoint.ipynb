# 1. Data Types

### 1.1 String and Number Data Tpyes

- Number data types consist of numbers, including integers (0,1,2,etc.), and floating point numbers with decimal places (0.1,0.12,etc.). They are referred to as <strong>"numberic types"</strong> or <strong>"number data types"</strong>.

- Numeric types can perform calculations such as addition, substraction, multiplication, and division.

- Words or sentences are in the form of a sequence of characters.
- They must be enclosed in single or double quotation marks to indicate to the computer where to start and end reading.
- The data type enclosed in quotation marks is called <strong>"sting data types"</strong> and is commonly referred to as a <strong>"string"</strong>
