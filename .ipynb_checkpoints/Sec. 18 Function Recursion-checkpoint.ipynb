### 1.2 Factorial and loops 

```python
n = int(input('Enter a number : '))
fact = 1 
for i in range(1, n + 1) :
    fact = fact * i
    
print('{}! = {}'.format(n, fact))

```

### 1.3 Factorial and recursion

- The recursion refers to a function that calls itself from within its own body.
- Recursion is a very useful problem-solving technique that allows us to solve complex problems in a intruitive and simple way, which would be difficult to solve using procedural techniques.

<span style="color:darkred">Focus:</span>  A termination condition is essential in a recursive function.
---

- Based on the definition, factorial can be defined as if '1! = 1', 'n! = n * (n - 1)!'.
- The important point in recursive functions is when the program terminates.
- According to the definition, if n is less than or equal to 1, we don't need to define factorial for values less than 1.
- Therefore, if the 'n' in the if statement becomes 1 or less, we can simply return 1.
- It can be implemented in recursive function code as follows:

```python
def factorial(n) :
    if n <= 1 :
        return 1
    else :
        return n * factorial(n - 1)
n = 5
print('{}! = {}'.format(n, factorial(n)))
```

## 2. Calculating Fibonacci sequence 


### 2.2 Fibonacci sequence and recursion 

```python
def fibonacci(n) :
    if n <= 1 :
        return n
    else :
        return(fibonacci(n - 1) + fibonacci(n - 2))
nterms = int(input("How many Fibonacci number do you want? "))

if nterms <= 0 :
    print("Error : Enter positive number. ")
else :
    print("Fibonacci Sequence : ", end = '')
    for i in range(nterms) :
        print(fibonacci(i), end='')
        
```

```python
def fibonacci(n) :
    if n == 0 :
        return 0
    elif n == 1 :
        return 1 
    else :
        return fibonacci(n - 1) + fibonacci(n - 2)
print(fibonacci(10))
```

Explaination:
- The value accumulates as it is recursively called until it meets the termination condition.

### 2.3 Efficiently of Fibonacci Sequence Function Implemented with Recursive Calls
- The biggest issue with recursive functions is their inefficiency because they unnecessarily repeat calculatioins even for values that have already been computed.
- Let's take the process of finding the 30th Fibonacci number as an example.

```python
from timeit import *

def fibonacci_1(n) :
    if n == 0 :
        return 0
    elif n == 1 :
        return 1
    else :
        return fibonacci_1(n - 1) + fibonacci_1(n - 2)
    
t3 = Timer("fibonacci_1(30)", "from__main__import fibonacci_1")
print("fibonacci_1(30) * 20 times :", t3.timeit(number = 20), "seconds")
```
