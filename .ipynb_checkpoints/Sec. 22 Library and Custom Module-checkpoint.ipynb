```python
import fund_cal

def main() :
    t = int(input('Enter the total number of people attending the hospital fund event: '))
    total = fund_cal.fund(t)
    
    print('Total fundraising is: ', total)
    
main()
```

Explanation in cal_n.py file

Line 1 :
- Algorithm to calculate the sum of numbers from 1 to the input value n.
 
Line 2 :
- Double slashes represent division operation.

Line 4 :
- Algorithm to calculate the sum of consecutive numbers from 1 to the input value n.

```python
import cal_n

def main() :
    n = int(input('Enter the number you want to calculate:' ))
    
    sum_v = cal_n.sum_n(n)
    print('The sum of all the numbers from 1 to', n, 'is:', sum_v)
    
    sum_vv = cal_n.sum_n2(n)
    print('The value of adding all the consecutive numbers from 1 to', n, 'is:', sum_vv)
    
main()
```
