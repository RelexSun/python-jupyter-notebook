### 2.3 Methods related to uppercase and lowercase

- <span style="color:orange;">The lower method</span> converts uppercase letters to lowercase in the string. Conversely, <span style="color:blue;">the upper method</span> converts lowercase letters to uppercase.

```python
s = 'hello, world'
s.lower()
```

```python
s.upper()
```

- <span style="color:orange;">The capitalize method</span> converts only the first charaacter to uppercase.

```python
s = 'hello'
s.capitalize()
```

- <span style="color:orange;">The swapcase method</span> is used to toggle the case of the characters in the string. In other words, it converts uppercase letters to lowercase letters to uppercase.

```python
'I Love Python'.swapcase()
```

- <span style="color:orange;">The istitle method</span> returns True or False indicating whether the string is in title case, i.e, whether the first letter of each word is capitalized and the remaining letters are lowercase.

```python
'I Love Python'.istitle()
```

### 2.4 startswith method to check the first character in a specific range

- startswith indicates whether the string starts with a specific character.

```python
s = 'ABCDEFGHI'
s.startswith('A')
```

```python
s = 'ABCDEFGHI'
s.startswith('B')
```

```python
s = 'ABCDEFGHI'
s.startswith('B', 1)
```

```python
s = 'ABCDEFGHI'
s.startswith('B', 5)
```

### 2.5 endswith method to check the last character in a specific range

- endswith indicates whether the string ends with a specific character.

```python
s = 'ABCDEFGHI'
s.endswith('I')
```
