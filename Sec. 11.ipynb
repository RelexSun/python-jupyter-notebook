## 5.1 pop method

- The pop method deletes a specific key-value pair and returns it.

```python
dic = {'a': 10, 'b': 20, 'c': 30, 'd': 40}
print(dic.pop('a'))
print(dic)
```

- pop(key) deletes a specific key-value pair from the dictionary and returns the deleted value.
- The following example deletes the key 'a' from the dictionary dic and returns 10.

#### The pop function has the key and the default value as a parameter.

- pop(key, default): When a default value is specified, if the key does not exist, it returns only the default value.

```python
print(dic.pop('f', 8))
```

> - Since the key 'f' does not exist in dic, it returns 8.

## 5.2 popitem method 

- popitem deletes the last or an arbitrary key-value pair from the dictioinary and returns the deleted pair as a tuple.

```python
dic = {'a': 10, 'b': 20, 'c': 30, 'd': 40}
print(dic.popitem())
```

> - The bahvior of popitem varies depending on the Python version. In Python 3.6 and above, it deletes the last-key-value pair, while in Python 3.5 and below, it deletes an arbitrary key-value pair. 

## 5.3 clear method 

- clear deletes all key-value pairs in the dictioinary.

```python
dic = {'a': 10, 'b': 20, 'c': 30, 'd': 40}
print(dic)
dic.clear()
print(dic)
```

## 5.4 update mathod

- When the keys in the dictionary are strings, you can use the update(key=value) method to add the value of the key to the dictionary.
- If the key does not exist in the dictioinary, the update method adds the key-value pair.

```python
dic = {'a': 10, 'b': 20, 'c': 30, 'd': 40}
dic.update(a = 90)
print(dic)
```

```python
dic = {'a': 10, 'b': 20, 'c': 30, 'd': 40}
dic.update(a = 900, f = 60)
print(dic)
```

- The form update(key=value) can only be used when the key is a string. If the key is a number, you can use updates(dictionary) by passing a dictionary to modify the value.

```python
dic2 = {1: 'one', 2: 'two'}
dic2.update({1: 'ONE', 3: 'THREE'}) 
print(dic2)
```

## 5.5 get method

```python
dic = {'a': 10, 'b': 20, 'c': 30, 'd': 40}
dic['e']
```

#### get(key) returns the value of a specific key in the dictionary.

```python
dic.get('e')
```

```python
dic.get('a')
```

- With get(key, default), if a default is specified, it returns the value of the key in the dictioinary, but if the key doesn't exist, it returns the default value.

```python
b = dic.get('z', 0)
print(b)
```

`In the dictionary dic, the key 'z' doesn't existm so it returns the default value of 0 that was specified.`

# 4 Let's code

```python
cafe_dic = {
    'Americano': 3000,
    'Iced Americano': 3500,
    'Cappuccino': 4000,
    'Caffe Latte': 4500,
    'Espresso': 3600,
}
print("      Cafe Menu")
for key in cafe_dic :
    print("{:16}: {:,}$".format(key, cafe_dic[key]))
print('=' * 40)


a = input()

if a in cafe_dic.keys() :
    print("{}: {}$".format(a, cafe_dic[a]))
else :
    print('item is not in the menu')
```
