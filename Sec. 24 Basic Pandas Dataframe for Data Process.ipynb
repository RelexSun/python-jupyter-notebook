## 2. Creating a DataFrame


```python
import pandas as pd 
data = {'col_1': [3, 2, 1, 0], 'col_2': ['a', 'b', 'c', 'd']}
pd.DataFrame(data) ## Convert the dictionary named 'data' into a DataFrame
```

```python
import pandas as pd
s = pd.Series([1.0, 2.0, 3.0], index = ['a', 'b', 'c']) ## Save the series object as 's'
s2 = pd.Series([1.0, 2.0, 3.0, 4.0],  index = ['a', 'b', 'c', 'd'])## Save the series object as 's2'

data = {'one': s, "two": s2} ## Load the two series and assign the key names as 'one' and 'two'.

## When printing the DataFrame, the key names from the loaded dictionary become the column names.
pd.DataFrame(data) 
```

```python
import pandas as pd
d = {'col1': [1, 2], 'col2': [3, 4]} ## It is dictionary data type.
df = pd.DataFrame(data = d) ## Create a DataFrame object named 'df' by assigning a dictionary.

## Let's check the DataFrame created before changing the names.
print(df)

df.index = ['new index0', 'new index1'] ## Change to new index names.
print(df) ## Check the result of the index change.

df.columns = ['new_col1', 'new_col2'] ## Change to new column names.
print(df) ## Check the result of the column name change.
```

```python
import pandas as pd
d = {'col1': [1, 2], 'col2': [3, 4]}
df = pd.DataFrame(data = d) 
print(df) ## Let's check the DataFrame created before changing the names.

df.rename(index = {0: 'new index0'}) ## Change index 0 to a new name, 'new index0'.
```

```python
df.rename(columns = {"col2": 'new col2'})
```

```python
import pandas as pd
d = {'col1': [1, 2], 'col2': [3, 4]}
df = pd.DataFrame(data = d) 

## Use 'inplace = True' to modify the original DataFrame object itself.
df.rename(index = {0: 'new index0'}, inplace = True)
## If 'inplace' is not set, the default value is 'inplace=False', so the original object
## itself is not changed
df.rename(columns = {"col2": 'new col2'})
```
