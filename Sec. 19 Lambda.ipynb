### 1.3 Converting a regular function to a lambda
- The code below shows the addition of integers using the add function.
- Despite being a simple function that adds two numbers, it still needs to follow all the syntaxes such as the 'def' keyword, function name, function body, and the 'return' keyword for returning.

```python
# function
def add(x, y) :
    return x + y

print('The sun of 100 and 200 :', add(100, 200))
```

```python
# lambda
add = lambda x, y : x + y
print('The sun of 100 and 200 :', add(100, 200))
```

- We define a lambda function with the keyword 'lambda', which takes x and y as parameters and has expression x + y as the return value.
- It is composed in a form that has the functionality of adding two arguments without a header such as 'def add(x,y):'.

```python
print('The sun of 100 and 200 :', (lambda x, y : x + y)(100, 200))
```

- The lambda function performs the functionality of x + y and returns it.
- It passes the arguments 100 and 200 to the parameters x and y.
