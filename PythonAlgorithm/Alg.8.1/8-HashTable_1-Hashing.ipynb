### Typically, r and M should be relatively prime numbers.

----

### Definition
- hashing: The process of converting an arbitrary key into a key of a predetermined length.
- hash function: An algorithm that converts a given key into a <span style="color: blue">hash key</span> of a predetermined length.
- hash collision: When two different keys are converted into the same <span style="color: blue">hash key</span>.

----

```python
def hash(s) :
    key = 0
    for i in range(len(s)) :
        key += (ord(s[i]) - ord('a') + 1) * (r ** i)
    return key % M

r, M = map(int, input().split())
S = input()
hashkey = hash(S)
print(hashkey)
```
