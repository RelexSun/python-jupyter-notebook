----

### Therefore, when positive integers a, b and m are given, the simplest stradegy to calculate a^b mod m is to calculate a^b and then find the remainder when divided by m.

----

```python
def solve(a, b, m) :
    r = a
    for _ in range(b - 1) :
        r = r * a
    return r % m

A, B, M = map(int,input().split())
print(solve(A, B, M))
```

However, this method can become very slow when performing multiplication with large values of a, as the result grows exponentially.

# Create Solution 2
- In the first solution, it was observed that multiplication of large positive integers takes a significant amount of time as an operation.

### (A x  B)mod M = {(A mod M) x (B mod M)}mod M
- When we multiply A and B, they can become very large numbers.
- However, by applying the mod(modular) operation to A and B seperately before multiplication, we reduce them to smaller values, reducing the number of multiplications of large integers.
- Therefore, this stradegy leads to significant performance improvements.

```python
def solve(a, b, m) :
    r = a
    for _ in range(b - 1) :
        r = r * a % m
    return r

A, B, M = map(int, input().split())
print(solve(A, B, M))
```
