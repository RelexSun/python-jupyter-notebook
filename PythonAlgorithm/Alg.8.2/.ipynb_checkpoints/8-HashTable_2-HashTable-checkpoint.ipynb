-----

## This method allows for much more <span style="color: orange">efficient storage and retrieval of data</span> compared to traditional sequential search or binary search algorithms.
- This is because the <span style="color: orange">hash function</span> determines which bucket the data is stored in, and this calculation is done in constant time, O(1).
- It is significantly faster than the O(n) of Sequential Search or the O(logn) of binary search algorithms.

-----

```python
from collections import defaultdict
class HashTable :
    def __init__(self, size) :
        self.size = size
        self.table = defaultdict(list)
    def hash(self, key) :
        return key % self.size
    def get(self, key) :
        return self.table[self.hash(key)]
    def put(self, key, value) :
        self.table[self.hash(key)].append(value)

from HashTable import HashTable

N, M = map(int, input().split())
hashTable = HashTable(M)
for i in range(N) :
    book = input()
    key = sum(map(ord, book))
    hashTable.put(key, book)
for key in range(M) :
    print(key, ", ".join(hashTable,get(key)))
```
