```python
def solve(x) :
    stack = []
    for x in X :
        if x.isdigit() :
            stack.append(int(x))
        else :
            b, a = stack.pop(), stack.pop()
            if x == '+' : stack.append(a+b) 
            elif x == '-' : stack.append(a-b)
            elif x == '*' : stack.append(a*b)
            elif x == '/' : stack.append(a//b)
    return stack[-1]

X = input()
print(solve(X))        
```
