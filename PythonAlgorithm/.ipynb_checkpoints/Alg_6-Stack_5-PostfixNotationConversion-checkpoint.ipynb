----

 - Traverse the infix notation from left to right.
 - If we <strong>encounter an operand</strong>, immediately move it to the postfix notation.
 > - Pop operators from the stack to the postfix notation until the priority of the current operator is higher than the operator at the top of the stack.
 > - Then, push the current operator onto the stack.
 
 - If we <strong>encounter an open parenthesis</strong>, push it onto the stack.
 > - Repeatedly pop operators from the stack to the postfix notation until we encounter an open parenthesis.
 > - The encountered open parenthesis is popped from the stack but not moved to the postfix notation.
 
 - When we have finished traversing the entire infix notation
 > - Move any remaining operators in the stack one by one to the postfix notation.

---

```python
precedence = {'(':0, ')':1, '+':2, '-':2, '*':3, '/':3}
def solve(X) :
    stack, postfix = [], ""
    for x in X :
        if x.isalpha() : postfix += x
        elif x == '(' : stack.append(x) 
        else :
            while stack and precedence[x] <= precedence[stack[-1]] :
                postfix += stack.pop()
            if x == ')': stack.pop() # remove '('
            else : stack.append(x) # operator
                
    while stack :
        postfix += stack.pop()
    return postfix

X = input()
print(solve(X))
```
