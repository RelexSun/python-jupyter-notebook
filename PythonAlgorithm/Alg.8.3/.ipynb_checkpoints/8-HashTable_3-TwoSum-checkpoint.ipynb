```python
def solve(n, k, s) :
    for i in range(n) :
        for j in range(i + 1, n) :
            if s[i] + s[j] == k :
                print(s[i], s[j])

N, K = map(int, input().split())
S = list(map(int, input().split()))
solve(N, K, S)
```

```python
def solve(n, k, s) :
    for i in range(n) :
        if k - s[i] in s and i < s.index(k - s[i]) :
            print(s[i], k - s[i])

N, K = map(int, input().split())
S = list(map(int, input().split()))
solve(N, K, S)
```

```python
def solve(n, k, s) :
    d = {s[i]:k-s[i] for i in range(n)}
    for i in range(n) :
        if d[s[i]] in d and i < s.index(d[s[i]]) :
            print(s[i], d[s[i]])

N, K = map(int, input().split())
S = list(map(int, input().split()))
solve(N, K, S)
```
