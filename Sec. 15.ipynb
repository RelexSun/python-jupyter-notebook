### 2.2 Defining a set


- To create an empty set, use the set function

```python
numbers = set()
numbers
```

- When creating a set, you can also use the set function with an already created list or tuple, as shown below.

```python
days_list = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
days_set = set(days_list)
days_set
```

```python
days_tuple = ('haha', 'lol', 'xd')
days_set = set(days_tuple)
days_set
```

- It is also possible to create a set from a `list` as shown below.
- However, in the set data type, if there are duplicate elements, they are automatically removed.

```python
set([1, 2, 3, 1, 2])
```

- It is also possible to create a set from a `string`.
- In this case, each character becomes a single element.

```python
set('abcdefg')
```

```python
h_str = 'hello'
h_set = set(h_str)
h_set
```

### 2.3 Membership operators for searching specific elements: in, not in

- To check if an item is in the set, use the `in` operator.

```python
numbers = {1, 2, 3}
if 1 in numbers :
    print("There is a 1 in the set")
```

- To check if an item is not in the set, use the `not in` operator.

```python
numbers = {2, 1, 3}
if 7 not in numbers :
    print("There is not a 7 in the set")
```

### 2.4 Difference between a set and a sequence data type.

- Sets are unordered collections of unique elements and do not allow duplicates.
    - Therefore, a set is not a sequence data type and do not have the index

---
<span style="color:darkred">Focus:</span>  Since set elements do not have the index, indexing and slicing oprations are meaningless.
---
---
- instead, you can use the `in` operator to check if a specific element is present in the set.

- You can add a single element to a set using the `add` method.

```python
numbers = {1, 2, 3}
numbers.add(4)
numbers
```

- When removing elements from a set, you can use the `remove` method.

```python
numbers.remove(4)
numbers
```

### 2.5 Accessing sets using loops with iterable data type

- Sets are iterable data types, so you can use loops to access their elements.
- You can iterate over each item in a set using a for statement and print them out.

```python
numbers = {2, 1, 3}
for x in numbers :
    print(x, end=" ")
```

### 2.6 Sorting with the sorted function

<span style="color:darkred">Focus:</span>  Since sets are unordered, the order of elements may differ from the input order.
---

- If you want to print the items in a sorted order, you can use the `sorted` function.

```python
numbers = {2, 1, 3}
for x in sorted(numbers) :
    print(x, end=" ")
```

- The `sorted` function returns a list data type, so the set data type is converted to a list.

```python
numbers = sorted(numbers)
numbers
```

### 2.7 Creating sets using one-liner loop expressions.

- You can create a set using a one-liner expression as follows.
  - <mark>Expression for variable in iterable or sequence data type</mark>

```python
a
```

### 2.8 Creating sets using one-liner expressions with loops and conditions

- Using a one-liner expression allows you to achieve the desired result with concise coding. 
- In a one-liner expression, the if statement is specified after the for statement.
    - <mark>Expression for variable in iterable or sequence data type if condition</mark>
    
-  The following code extracts the characters from the 'pineapple' string excluding the alphabet 'a', 'p', 'l'. 

```python
a = {i for i in 'pineapple' if i not in 'apl'}
```

```python
a
```

> - Each alphabet value of 'pineapple' is assigned to i through the for statement.
> - This i value is filtered by the condition of the if statement.
    If i is not 'a', 'p', or 'l', it becomes True, and 'e', 'i', 'n' is stored in 'a'.

- There are logical functions, such as `all` and `any`, that can be applied to data with multiple items.

```python
a_set = {1, 0, 2, 3, 3}
all(a_set), any(a_set)
```

- `all` returns true if all the iterable data items inside it eveluate to true.
> - If any of them are false, it returns false.

- `any` returns true if at least one of the iterable data items inside it evaluates to true, regardless of the other values.
> - If all of them are false, it returns false.
